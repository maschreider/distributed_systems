```python
# listener
import hazelcast
from datetime import datetime

time_start = datetime.now().timestamp()

hz = hazelcast.HazelcastClient(cluster_name='task2') 
topic = hz.get_topic('my-distributed-topic').blocking() 

def print_message(topic_message):
    print("Recieved:", topic_message.message)

topic.add_listener(print_message)

while (datetime.now().timestamp() - time_start) < 50: 
    pass
else:
    hz.shutdown()
```

```python
# reader
import hazelcast

hz = hazelcast.HazelcastClient(cluster_name='task2') 
queue = hz.get_queue("my-distributed-queue").blocking()

item = 0

while item != -1:
    item = queue.take()
    print(f"Taken: {item}")
else:
    queue.put(-1)
    hz.shutdown()
```
