```python
import hazelcast
import time

hz = hazelcast.HazelcastClient(cluster_name='task2')
```

```python
map = hz.get_map('my-distributed-map').blocking() 
for key in range(1000): map.set(key, key)
```

```python
# publisher
topic = hz.get_topic('my-distributed-topic').blocking() 
for n in range(100): 
    topic.publish(n)
    # time.sleep(0.1)
```

```python
# writer
queue = hz.get_queue("my-distributed-queue").blocking()
for n in range(100): 
    queue.put(n)
    print(f'n: {n} | size: {queue.size()} | capacity: {queue.remaining_capacity()}')
queue.put(-1)
hz.shutdown()
```
